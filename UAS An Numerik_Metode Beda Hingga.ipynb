```python
from numpy import *
from numpy.linalg import *
from matplotlib.pyplot import *
```

```python
#t in [0.2], dt=0.1
dt = 0.1
a  = 0
b  = 2

N  = int( (b-a)/dt )
```

```python
t  = linspace(a,b,N+1)
x  = zeros(N+1)
y  = zeros(N+1)

x[0] = 8
y[0] = 5

A = [[1+4*dt,-5*dt],[2*dt,1-3*dt]]
X = [[x[0]],[y[0]]]

A = array(A)
X = array(X)

for n in range(N):
    X = dot(A,X)
    x[n+1] = X[0,0]
    y[n+1] = X[1,0]
```

```python
print('A =',A)
print('X =',X)
```

```python
def f(x):
    return 3*exp(-t)+5*exp(2*t)
def g(x):
    return 3*exp(-t)+2*exp(2*t)
```

```python
print('f(x)=',f(x))
print('g(x)=',g(x))
```

```python
plot(t,x,'r',t,f(t),'b')
show()
```

```python
plot(t,y,'r',t,g(t),'b')
show()
```
